```python
def get_image_description(data, title):
    text = str(data.plain_text)
    text = text.replace("\\","")
    text = text.replace("\'","")
    title_replaced = title.replace("\'","")
    num = text.find("[[File:"+title_replaced)
    chars = [char for char in text[num:]]
    esc_mode = False
    count = 0
    for idx,char in enumerate(chars):
        if char == '[':
            esc_mode = True
            count = 0
        if not esc_mode:
            if char == ']':
                bracket_pos = idx
                break
        else:
            if char == ']':
                count += 1
        if count == 2:
            esc_mode = False
            count = 0
    chars = [char for char in text[num:num+bracket_pos]]
    for idx, char in enumerate(chars):
        if char == '|':
            line_pos = idx
    raw = text[num+line_pos+1:num+bracket_pos]
    raw = raw.replace('[[', '')
    output = raw.replace(']]', '')
    return s2t(output)

for title in image_titles:
    try:
        print(get_image_description(data, title))
    except Exception as e:
        print(e)
```

```python
wiki = wikipediaapi.Wikipedia('zh')
page = wiki.page('宋美齡')
```

```python
def get_sections_list(page):
    sections_content = []
    sections_content.append(('',s2t(page.summary)))
    for idx,section in enumerate(page.sections):
        text_arr = [section.text]    
        if len(section.sections) != 0:
            for idx2, subsection in enumerate(section.sections):
                text_arr.append(subsection.text)
                for idx3, subsubsection in enumerate(subsection.sections):
                    text_arr.append(subsubsection.text)
        text = '\n'.join(text_arr)
        if not text.isspace() and not text=='':
            sections_content.append((s2t(section.title),s2t(text)))
    return sections_content
    
```

```python
sections_content = get_sections_list(page)
```
